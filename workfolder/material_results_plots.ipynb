```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

# Force white background
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'

# Updated dataset with more spread
data = {
    0:  [28.0, 31.5, 29.0, 32.0, 27.5],
    20: [13.0, 17.2, 15.0, 16.3, 12.8],
    40: [17.0, 22.1, 19.2, 21.4, 18.3],
    60: [22.5, 27.0, 24.0, 26.5, 23.3],
}

# Convert to DataFrame
df = pd.DataFrame([(p, val) for p, vals in data.items() for val in vals],
                  columns=['Dredged Sediment (%)', 'Failure Force (kN)'])

# Plot
fig, ax = plt.subplots(figsize=(8, 6))

for p, group in data.items():
    x_vals = [p] * len(group)
    ax.scatter(x_vals, group, s=60, label=f'{p}%')

# Mean trend line
means = [sum(vals)/len(vals) for vals in data.values()]
ax.plot(data.keys(), means, linestyle='--', color='black', marker='o', label='Mean Failure Force')

# Watermark
ax.text(30, 22, 'example', fontsize=48, color='grey', alpha=0.2,
        ha='center', va='center', rotation=0)

# Plot formatting
ax.set_title('Failure Force of 10×10×10 cm Cement Blocks vs. Dredged Sediment Content', fontsize=14)
ax.set_xlabel('Dredged Sediment Content (%)', fontsize=12)
ax.set_ylabel('Failure Force (kN)', fontsize=12)
ax.set_xticks([0, 20, 40, 60])
ax.tick_params(labelsize=11)
ax.legend(fontsize=10)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
